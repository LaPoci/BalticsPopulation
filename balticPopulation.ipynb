# Importuojamos bibliotekos

```python
from matplotlib import pyplot as plt
import pandas as pd

#pagalbine komanda
%matplotlib inline
```

# Baltijos šalių pavadinimų sąrašas (1 užduotis)

```python
baltic_countries = ["Lituania", "Estonia", "Latvia"]
baltic_countries
```

# Baltijos šalių sostinių pavadinimų sąrašas (2 užduotis)

```python
baltic_capitals = ["Vilnius", "Tallinn", "Riga"]
baltic_capitals
```

# Baltijos šalių gyventojų skaičius, žodynas (3 užduotis)

```python
baltic_population = {
    "Lithuania": 2.79,
    "Estonia": 1.33,
    "Latvia": 1.90
}
```

# Baltijos šalių sostinių gyventojų skaičius, žodynas (4 užduotis)

```python
baltic_capitals_population = {
    "Vilnius": 544386,
    "Tallinn": 426538,
    "Riga": 632614
}
```

# Baltijos šalių gyventojų skaičius, stulpelinė diagrama (5 užduotis)

```python
# Data
baltic_country_list = baltic_population.keys()
baltic_population_list = baltic_population.values()

# Bar chart
plt.figure(figsize=(14,6))

plt.bar(baltic_country_list, baltic_population_list, color = "royalblue")

plt.title("\nBaltijos šalių gyventojų skaičius, mln.\n", fontsize = 22)
plt.ylabel("\ngyventojai, mln.\n", fontsize=18)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.grid()
plt.show()
```

# Baltijos šalių sostinių gyventojų skaičius, stulpelinė diagrama (6 užduotis)

```python
# Data
baltic_capitals_list = baltic_capitals_population.keys()
baltic_population_list = baltic_capitals_population.values()

# Bar chart
plt.figure(figsize=(14,6))

plt.bar(baltic_country_list, baltic_population_list, color = "cyan")

plt.title("\nBaltijos šalių sostinių gyventojų skaičius\n", fontsize = 22)
plt.ylabel("\ngyventojai\n", fontsize=18)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.grid()
plt.show()
```

# Baltijos šalių gyventojų proporcija, skritulinė diagrama (7 užduotis)

```python
plt.figure(figsize=(12, 6))
plt.pie(baltic_population_list, labels = baltic_country_list, autopct = "%1.0f%%", textprops = {"fontsize": 14})

plt.title("\nBaltijos šalių gyventojų pasiskirstymas\n", fontsize = 22)
plt.show()
```

# Baltijos šalių gyventojų skaičiaus dinamika 2015 - 2020 metais (8 užduotis)

```python
df_baltic_population = pd.read_csv("baltic_population.csv", index_col = "Year")
df_baltic_population
```

```python
fig, ax = plt.subplots(figsize = (12, 6))

df_baltic_population.plot(ax = ax)
plt.title("\nBaltijos gyventojų skaičiaus dinamika\n", fontsize = 22)
plt.xlabel("Year", fontsize = 18)
plt.ylabel("\nGyventojai\n", fontsize = 18)

plt.legend(loc = 'upper right', fontsize = 14, bbox_to_anchor = (1.5, 1))

plt.grid()
plt.show()
```

# Baltijos gyventojų skaičiaus dinamika 2015 - 2020 metais, grouped bar chart (9 užduotis)

```python
fig, ax = plt.subplots(figsize=(12, 6))
df_baltic_population.plot.bar(ax=ax)

plt.title("\nBaltijos gyventojų skaičiaus dinamika\n", fontsize = 22)
plt.xlabel("Year", fontsize = 18)
plt.ylabel("Gyventojai", fontsize = 18)

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)

plt.legend(loc = "upper right", fontsize = 18, bbox_to_anchor = (1.4, 1))

plt.grid()
plt.show()
```
